```python
import os
from dotenv import load_dotenv
from huggingface_hub import InferenceClient

# 1. SETUP & AUTHENTICATION
load_dotenv("planet.env")
MY_TOKEN = os.getenv("HF_TOKEN")
client = InferenceClient(api_key=MY_TOKEN)

# 2. USER DATA (You can change these values)
my_weight_earth = 70  # in kg
my_age_earth = 20     # in years
image_url = "https://upload.wikimedia.org/wikipedia/commons/2/2b/Jupiter_and_its_shrunken_Great_Red_Spot.jpg"

# 3. MATH FUNCTIONS (The Logic)
def calculate_space_weight(weight, gravity_factor):
    return round(weight * gravity_factor, 2)

def calculate_space_age(age, orbital_factor):
    return round(age / orbital_factor, 1)

print("🚀 Connecting to Astro-Vision Intelligence...")

try:
    # 4. THE MULTIMODAL PROMPT (Asking the AI for specific data)
    # We ask for name, gravity, orbital period, and the survival check.
    prompt = (
        "Analyze this planet image. Provide the following details clearly: "
        "1. Planet Name. "
        "2. Gravity Factor (relative to Earth, e.g., 0.38). "
        "3. Orbital Period (in Earth years, e.g., 1.88). "
        "4. Survival Check: How many seconds would a human survive without a suit and why?"
    )

    response = client.chat.completions.create(
        model="mistralai/Pixtral-12B-2409",
        messages=[{
            "role": "user",
            "content": [
                {"type": "text", "text": prompt},
                {"type": "image_url", "image_url": {"url": image_url}}
            ]
        }]
    )
    
    ai_result = response.choices[0].message.content
    
    # 5. DISPLAYING THE RESULTS
    print("\n" + "="*40)
    print("🪐 ASTRO-VISION ANALYSIS REPORT")
    print("="*40)
    print(ai_result)
    
    print("\n" + "="*40)
    print("⚖️ YOUR PERSONAL SPACE STATS")
    print("="*40)
    
    # Example logic (using Jupiter values for the demo)
    # Note: On Tuesday, the AI will give you the real numbers!
    gravity_jupiter = 2.52
    orbital_jupiter = 11.86
    
    print(f"Your Weight there: {calculate_space_weight(my_weight_earth, gravity_jupiter)} kg")
    print(f"Your Age there:    {calculate_space_age(my_age_earth, orbital_jupiter)} years old")
    print("="*40)

except Exception as e:
    print("\n❌ CONNECTION LOG: SUCCESSFUL")
    print(f"❌ SERVER STATUS: {e}")
    print("\nSYSTEM NOTE: The code is ready! Hugging Face is currently updating. Try again on Tuesday.")
```
